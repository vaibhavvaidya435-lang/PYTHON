# Types of Operator

```python
#Arithmetic Operators (+ , - , * , / , % , // , **)

a = 5
b = 2
print(a + b) #Addition
print(a - b) #Subtraction
print(a * b) #Multiplication
print(a / b) #Division
print(a % b) #Modulus #remainder
print(a ** b) #Exponentiation #power
```

```python
#Relational Operators (== , != , > , < , >= , <=) give boolean output

a = 50
b = 20

print(a == b) #Equal to
print(a != b) #Not Equal to
print(a > b)  #Greater than
print(a < b)  #Less than
print(a >= b) #Greater than or Equal to
print(a <= b) #Less than or Equal to    
```

```python
#Assignment Operators(= , += , -= , *= , /= , %= , **=)

num = 10 
num = num + 10 # also use num += 10 # 10 + 10 = 20
print ("num : ", num)
```

```python
#logical Operators (and , or , not)

a = 50
b = 30
print(not False) #True
print(not(a>b)) #False
#also
val1 = False
val2 = False
print("AND operator :", val1 and val2)# and operator gives True if both values are True else False
print("OR operator :",(a == b)or (a > b))# or operator gives True if any one value is True else False
```
