```python
#indexing
str1="hello"
print(str1[0])
str2="vaibhav"
print(str2[3])
print(str2[-1])  #negative indexing
print(str2[-4])
print(str2[0:4])  #slicing
print(str2[1:5])
print(str2[:4])
print(str2[2:])
print(str2[:])  #full string
print(str2[-5:-1])
print(str2[-4:])
print(str2[:-2])
print(str2[-6:])
print(str2[: :-1])  #reversing the string
print(str2[::2])  #skiping one character
print(str2[1::2])  #skiping one character from index 1
print(str2[::-2])  #reversing and skiping one character
print(str2[5:1:-1])  #slicing in reverse order
print(str2[4:0:-1])  #slicing in reverse order
print(str2[::])  #full string
print(str2[-1::-1])  #full string in reverse order
print(str2[-1: -len(str2)-1 : -1])  #full string in reverse order
```

```python
# string functions
str1="  hello world  "
print(str1.upper())
print(str1.lower())
print(str1.strip())  #removes spaces from start and end
print(str1.replace("world", "vaibhav"))
print(str1.split())  #splits the string into list of words
print(str1.find("world"))  #gives starting index of substring
print(str1.count("l"))  #counts number of occurrences of a character
print(str1.startswith("  he"))  #checks if string starts with given substring
print(str1.endswith("ld  "))  #checks if string ends with given substring
print(str1.isdigit())  #checks if string is digit
print(str1.isalpha())  #checks if string is alphabetic
print(str1.isalnum())  #checks if string is alphanumeric
```

```python
#nesting
num=int(input("enter a number: "))
if num>0:
    if num%2==0:
        print("positive even number")
    else:
        print("positive odd number")
else:
    if num%2==0:
        print("negative even number")
    else:
        print("negative odd number")
```
