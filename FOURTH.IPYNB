```python
#dictionnary in python
my_dict = {
    "name": "vaibhav",
    "age": 18,
    "city": "Nashik"
}
print(my_dict)
```

```python
#nested dictionnary
my_dict2 = {
    "person1": {
        "name": "vaibhav",
        "age": 18
    },
    "person2": {
        "name": "rahul",
        "age": 20
    }
}
print(my_dict2)
```

```python
#methods in dictionnary
print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())
my_dict.update({"age": 19})
print(my_dict)
my_dict.pop("city")
print(my_dict)
print(len(my_dict))
print(my_dict.get("name"))
```

```python
my_dict.keys()#returns all keys
my_dict.values()#returns all values
my_dict.items()#returns all key-pair of tuples
my_dict.get("key")#returns the key according to value
my_dict.update({"key": "value"})#inserts or updates the key-value pair
my_dict.pop("key") #removes the key-value pair
            
```

```python
#set in python
my_set = {1, 2, 3, 4, 5}
print(my_set)
my_set.add(6)
print(my_set)
my_set.remove(3)
print(my_set)
print(len(my_set))
my_set.clear()
print(my_set)
```

```python
collection={1,2,3,4,5,5,5,5}#sets do not allow duplicate values
print(collection)
print(type(collection))
print(len(collection))
collection.add(6)#adds an element to the set
print(collection)
collection.remove(2)#removes an element from the set
print(collection)
print(3 in collection)#checks if an element is in the set
collection.clear()#clears the set
print(collection)
```

```python
#set methods
my_set = {1, 2, 3, 4, 5}
my_set.add(7)#adds an element to the set
my_set.remove(4)#removes an element from the set
my_set.clear()#clears the set
print(my_set)
print(len(my_set))
print(5 in my_set)#checks if an element is in the set
```

```python
set.union()#returns a set that is the union of two sets
set.intersection()#returns a set that is the intersection of two sets
set.difference()#returns a set that is the difference of two sets
set.symmetric_difference()#returns a set that is the symmetric difference of two sets
set.issubset()#checks if a set is a subset of another set
set.issuperset()#checks if a set is a superset of another set
set.copy()#returns a shallow copy of the set
set.pop()#removes and returns an arbitrary set element
set.update()#updates the set with the union of itself and others
```
